Here's the 2D line plot visualized in 3D space. The `z` coordinate is kept constant (at `0`) to show how a 2D line appears when plotted on a plane within a 3D environment.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Reusing sample data from the previous plot
# x = [1, 2, 3, 4, 5]
# y = [10, 20, 25, 30, 40]

# Create a constant z-axis for the 2D line in 3D space
z = np.zeros_like(x)

# Create a new figure and a 3D subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D line
ax.plot(x, y, z, color='green', linestyle='--', marker='s', markersize=8, linewidth=2)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('2D Line Plot in 3D Space')

# Add a grid for better readability
ax.grid(True)

# Show the plot
plt.show()
```

In this plot:

*   **`color='green'`**: Sets the line color to green.
*   **`linestyle='--'`**: Creates a dashed line.
*   **`marker='s'`**: Uses square markers for each data point.
*   **`markersize=8`**: Sets the size of the markers.
*   **`linewidth=2`**: Sets the thickness of the line.
*   **`plt.grid(True)`**: Adds a grid to the plot, which can help in reading values.

To normalize the confusion matrix to show proportions along the true labels, we can use the `normalize='true'` parameter in `ConfusionMatrixDisplay.from_predictions`. This will display the percentage of each class's predictions relative to the actual number of instances for that class.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Reusing y_true and y_pred from the previous cell
# y_true = [0, 1, 0, 1, 0, 1]
# y_pred = [0, 1, 0, 0, 0, 1]

# Create and display the normalized confusion matrix
disp_normalized = ConfusionMatrixDisplay.from_predictions(
    y_true=y_true,
    y_pred=y_pred,
    normalize='true',
    cmap=plt.cm.Blues,
    values_format='.2f' # Format values to two decimal places for percentages
)

plt.title('Normalized Confusion Matrix (Proportions by True Label)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate some random 3D data points
n = 100 # Number of points
x_data = np.random.rand(n)
y_data = np.random.rand(n)
z_data = np.random.rand(n)

# Create a new figure and a 3D subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter points
ax.scatter(x_data, y_data, z_data, c='r', marker='o')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot of Random Data')

# Show the plot
plt.show()
```

This 3D scatter plot demonstrates how to visualize individual points in three-dimensional space.

*   **`np.random.rand(n)`**: Generates `n` random numbers between 0 and 1 for each of the x, y, and z coordinates.
*   **`ax.scatter(x_data, y_data, z_data, c='r', marker='o')`**: This function is similar to `plt.scatter` but for 3D. It takes the three arrays of coordinates and plots them. `c='r'` sets the color to red, and `marker='o'` uses circular markers.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid from x and y arrays
X, Y = np.meshgrid(x, y)

# Calculate Z using the function Z = sin(sqrt(x^2 + y^2))
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Create a new figure for the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot of Z = sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

Here's an explanation of the code:

*   **`import numpy as np`**: Imports the NumPy library, which is essential for numerical operations, especially array creation and mathematical functions.
*   **`import matplotlib.pyplot as plt`**: Imports the standard plotting library.
*   **`from mpl_toolkits.mplot3d import Axes3D`**: This is crucial for 3D plotting. It imports the `Axes3D` object, which enables 3D axes within a Matplotlib figure.
*   **`x = np.linspace(-5, 5, 100)`** and **`y = np.linspace(-5, 5, 100)`**: These lines create one-dimensional arrays for `x` and `y` values, ranging from -5 to 5 with 100 points each. `linspace` is great for creating evenly spaced numbers.
*   **`X, Y = np.meshgrid(x, y)`**: This function creates 2D coordinate arrays from the 1D `x` and `y` arrays. `X` and `Y` will be 100x100 matrices where `X` holds the x-coordinates and `Y` holds the y-coordinates for every point in the grid.
*   **`R = np.sqrt(X**2 + Y**2)`** and **`Z = np.sin(R)`**: These lines calculate the `R` (radius from the origin) and then the `Z` value for each corresponding `(X, Y)` pair, applying the `sin(sqrt(x^2 + y^2))` function.
*   **`fig = plt.figure(...)`** and **`ax = fig.add_subplot(111, projection='3d')`**: These lines create a new Matplotlib figure and then add a 3D subplot to it. The `projection='3d'` argument is what makes the subplot 3D.
*   **`ax.plot_surface(X, Y, Z, cmap='viridis')`**: This is the core command for creating the 3D surface plot. It takes the `X`, `Y`, and `Z` meshgrid data. `cmap='viridis'` sets the colormap for the surface, adding visual appeal.
*   **`ax.set_xlabel(...)`, `ax.set_ylabel(...)`, `ax.set_zlabel(...)`, `ax.set_title(...)`**: These lines set the labels for the three axes and the title of the 3D plot.
*   **`plt.show()`**: Displays the generated 3D plot.

```python
import matplotlib.pyplot as plt

labels = ['Class 0', 'Class 1']
sizes = [4, 6]  # Example counts

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Class Proportion")
plt.show()
```

Here's a 3D bar chart representing the class proportions. While a pie chart is often sufficient for proportions, this demonstrates how to visualize counts in a pseudo-3D manner using bars.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Reuse labels and sizes from the previous cell
# labels = ['Class 0', 'Class 1']
# sizes = [4, 6]

# Create a figure and a 3D subplot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Positions for the bars on the x-axis
x_pos = np.arange(len(labels))

# For a simple 3D bar chart, we can set y to a constant value or distribute them
y_pos = np.zeros_like(x_pos)

# Depth of the bars
dx = dy = 0.8

# Heights of the bars (from sizes)
dz = np.array(sizes)

# Plot the 3D bars
ax.bar3d(x=x_pos, y=y_pos, z=np.zeros_like(x_pos), dx=dx, dy=dy, dz=dz, shade=True, color=['skyblue', 'lightcoral'])

# Set labels and title
ax.set_xlabel('Classes')
ax.set_ylabel('Depth (arbitrary for proportions)')
ax.set_zlabel('Count / Size')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_title('3D Bar Chart of Class Proportions')

plt.show()
```
