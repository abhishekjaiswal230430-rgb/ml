```python
# func10 - Find factorial

def func10(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact

print(func10(5))
```

```python
# func11 - Print fibonacci series

def func11(n):
    a = 0
    b = 1

    for i in range(n):
        print(a, end=" ")

        c = a + b
        a = b
        b = c

func11(10)
```

```python
# func12 - Reverse a string

def func12(text):
    return text[::-1]

print(func12("Python"))
```

```python
# func13 - Check palindrome

def func13(text):

    if text == text[::-1]:
        return True
    else:
        return False

print(func13("madam"))
```

```python
# func14 - Count vowels in string

def func14(text):

    vowels = "aeiouAEIOU"
    count = 0

    for ch in text:
        if ch in vowels:
            count += 1

    return count

print(func14("Programming"))
```

```python
# func18 - Check prime number

def func18(num):

    if num < 2:
        return False

    for i in range(2, num):

        if num % i == 0:
            return False

    return True

print(func18(11))
```

```python
# func19 - Swap two numbers

def func19(a, b):

    a, b = b, a

    return a, b

print(func19(5, 10))
```

```python
# func21 - Convert celsius to fahrenheit

def func21(c):

    fahrenheit = (c * 9/5) + 32

    return fahrenheit

print(func21(37))
```

```python
# func22 - Find area of circle

def func22(radius):

    pi = 3.14

    area = pi * radius * radius

    return area

print(func22(5))
```

```python
# func23 - Calculate simple interest

def func23(p, r, t):

    si = (p * r * t) / 100

    return si

print(func23(1000, 5, 2))
```

```python
# func24 - Print multiplication table

def func24(num):

    for i in range(1, 11):
        print(num, "x", i, "=", num * i)

func24(5)
```

```python
# func25 - Check leap year

def func25(year):

    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True

    return False

print(func25(2024))
```
